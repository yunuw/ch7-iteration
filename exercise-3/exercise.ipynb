Open the `LICENSE` file (remember to use the `with` keyword), then use a `for` loop to print its contents (specify an `end` keyword argument when printing to avoid outputting extra newlines).

```python
with open('LICENSE') as license:
    for line in license:
        print(line, end="")
```

Open the `.gitignore` file, then use a `for` loop to count the number of entries in the file.
Ignoring any lines that [start with](https://docs.python.org/3.6/library/stdtypes.html#str.startswith) `#` (comments) or are blank (just a newline `\n`).

```python
with open('.gitignore') as gitignore:
    count = 0
    for line in gitignore:
        if not line.startswith('#') and not line == "\n":
            count = count + 1
            #print(line, end="")  # for testing
    print(count)
```
